## 1. Description


This repository is for lab work for the course Geospatial Analysis using Python. First you have to create conda environment as this lab work will use multiple Python packages to work with raster data. To set up a local `conda` environment, download [this conda environment definition](https://raw.githubusercontent.com/geohackweek/tutorial_contents/master/raster/environment.yml) and create an environment like so:

Or if you have downloaded the repository which containes the template for the lab work of the course then proceed as follows: 

- Open the command prompt
- change your working directory to the downloaded repository
 > <b><i>  cd  &lt; Path to the folder containing the notebooks &gt; </i></b>

- Run the following commands in the command prompt:
> <b><i> conda env create -f env_pyee01.yml </i></b>                
> <b><i> activate pyee01</i></b>
